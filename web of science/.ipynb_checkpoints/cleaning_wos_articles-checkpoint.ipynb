```python
import pandas as pd
import ast
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
wos = pd.read_csv("datasets/wos/authors_data_wos.csv")
docs = pd.read_csv("datasets/wos/docs_info.csv")
jour = pd.read_csv("datasets/wos/journal_info.csv")
```

```python
df_journal = jour[['title','issn','thematic_scope',]]
jour.columns
```

```python
docs.columns
```

```python
wos = wos[['Articles','Journal']]
wos['Articles'] = wos['Articles'].apply(ast.literal_eval)
wos['Journal'] = wos['Journal'].apply(ast.literal_eval)
```

```python
# Charger les données dans un DataFrame
df = pd.DataFrame(wos['Articles'])

# Décompresser les listes en lignes
df_exploded = df.explode('Articles', ignore_index=True)

# Transformer les dictionnaires en colonnes
df_articles = pd.json_normalize(df_exploded['Articles'])

# Afficher le résultat
df_articles
```

```python
# Charger les données dans un DataFrame
df_j = pd.DataFrame(wos['Journal'])

# Décompresser les listes en lignes
df_exploded_j = df_j.explode('Journal', ignore_index=True)

# Transformer les dictionnaires en colonnes
df_journal = pd.json_normalize(df_exploded_j['Journal'])

# Afficher le résultat
df_journal
```

```python
df_combined = pd.concat([df_articles,df_journal], axis=1)
df_combined = df_combined[['Titre de l’article','Mots-clés','Résumé','issn','scope']]
```

```python
df_combined = df_combined.drop_duplicates()
```

```python
df_combined = df_combined.rename(scope='Titre de journal')
```
