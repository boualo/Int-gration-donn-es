```python
import pandas as pd
```

```python
wos = pd.read_csv("web of science/wos_articles.csv")
scol = pd.read_csv("datasets/Articles_Dataset.csv")
scop = pd.read_csv("datasets/Training_data.csv")
```

```python
# dfsu.to_csv('wos_articles.csv',index=False)
# Filtrer les lignes oÃ¹ 'sjr_score' n'est pas Ã©gal Ã  'N/A' et n'est pas NaN
filtered_df = scop[(scop['Score SJR'] != 'N/A') & (scop['Score SJR'].notna())]

# Convertir 'sjr_score' en float aprÃ¨s filtrage
filtered_df['Score SJR'] = filtered_df['Score SJR'].astype(float)

# Appliquer la condition pour sjr_score > 1
result_df = filtered_df[filtered_df['Score SJR'] > 1]

(result_df)
```

```python
wos.columns , scol.columns , scop.columns
```

```python
# Renommer les dataFrames
wos.columns = ['title', 'author_keywords', 'abstract', 'issn', 'journal_name','sjr_score']

scol.columns = ['title', 'journal_name', 'abstract', 'issn','sjr_score']

scop.columns = ['issn', 'title', 'abstract', 'author_keywords', 'journal_name', 'sjr_score']
```

```python
# Combiner les dataFrames en un seul dataFrame
df_merged = pd.concat([wos,scop,scol],ignore_index=True)
df_merged
```

```python
df_merged.duplicated().sum()
```

```python
df_merged.drop_duplicates(inplace=True)
```

```python
df_merged.isna().sum()
```

```python
# ComplÃ©ter les valeurs manquantes en prenant celles du mÃªme titre
df_merged['author_keywords'] = df_merged.groupby('title')['author_keywords'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df_merged['sjr_score'] = df_merged.groupby('issn')['sjr_score'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df_merged['abstract'] = df_merged.groupby('title')['abstract'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df_merged
```
